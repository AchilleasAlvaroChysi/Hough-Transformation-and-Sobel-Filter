Σε αυτή την εργασία δημιουργήσαμε τον μετασχηματισμό Hough για τον εντοπισμό κύκλων σε φωτογραφίες. Για τον εντοπισμό χρησιμοποιήσαμε το φίλτρο Sobel που υλοποιήσαμε στην άσκηση 2. Στη συνέχεια υλοποιήσαμε ένα σύστημα ψηφοφορίας με κάδους. Επειδή κάνουμε αναζήτηση για κάθε pixel και κάθε ακτίνα κύκλου η καταχώρηση ψήφων γίνεται σε ένα 3Δ πίνακα όπου η πρώτη διάσταση είναι η συντεταγμέμη x η δεύτερη η συντεταγμένη y και η τρίτη η ακτίνα r. Αρχίκα περνάμε την εικόνα από το φίλτρο Sobel και στη συνέχεια για κάθε pixel που ανήκει σε ακμή (η τιμή του Sobel είναι 255) και κάθε ακτίνα δίνουμε μία ψήφο σε κάθε pixel γύρω του σε απόσταση r. Αυτό το κάνουμε κάνοντας ένα κύκλο από 0 - 360 μοίρες και παίρνοντας τις πολικές συντεταγμένες κάθε pixel. Μετά τη ψηφοφορία βρίσκουμε τα τοπικά ελάχιστα που ξεπερνάνε μία τιμή και κρατάμε τον κύκλο αυτό. Για να αποφύγουμε ομόκεντρους κύκλους ή κύκλους με κοντινά κέντρα, αν εντοπίσουμε ένα κύκλο ο οποίος πέφτει μέσα σε ενα κύκλο που έχουμε εντοπίσει ήδη τότε κρατάμε τον κύκλο με τις περισσότερες ψήφους. Τέλος ζωγραφίζουμε τους κύκλους πάνω στην εικόνα, την αποθηκεύουμε και μετά την ζωγραφίζουμε στην οθόνη. Μπορείτε να τρέξετε το πρόγραμμα χρησιμοποιόντας python 3 και δίνοντας σαν όρισμα μία εικόνα. Στην γενική περίπτωση το πρόγραμμα ψάχνει για κύκλους μήκους 10 με 20 pixels όπως ζητήται από την εκφώνηση. Μπορείτε όμως να δώσετε δύο επιπλέον ορίσματα με τις ακτίνες σε pixel.
